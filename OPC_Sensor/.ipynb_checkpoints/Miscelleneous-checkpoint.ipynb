# Errors Plot

```python
#FOR OPC
import matplotlib.pyplot as plt
import numpy as np
import math
X = ["MSE","MAE","RMSE"]
xg = [9.53231e-06, 0.0008703509, 0.003087444]
rf = [5.4943812238677005e-05,0.0021928648657524178, 0.0074124093410089685]
lasso = [0.00013819191, 0.0043099066, 0.011755506]
knn = [4.8233207e-05, 0.0018312254, 0.006945013]
plt.plot(X, xg, color='r', label='xgboost')
plt.plot(X, rf, color='g', label='randomforest')
plt.plot(X, knn, color='b', label='knn')
plt.plot(X, lasso, color='m', label='lasso')

plt.xlabel("Error")
plt.ylabel("Values")
plt.title("Plot for mae,mse and rmse for different models")
plt.legend()
plt.show()
```

```python
#FOR PJ

import numpy as np
import matplotlib.pyplot as plt

X = ['XGBOOST','RANDOMFOREST','LINEAR','KNN']
Train = [0.7867491776691122,0.6329507787480243,0.2929796878382063,0.7463683061096252]
Test = [0.804894044676069,0.6406985506761919,0.2948116798857159,0.8164847956978979]

xgboost_list = [0.7867491776691122, 0.804894044676069]
randomforest_list = [0.6329507787480243, 0.6406985506761919]
Linear_regression_list = [0.2929796878382063, 0.2948116798857159]
Lasso_list = [-8.30655766355311e-06, 3.407978172024928e-12]
knn_list = [0.7463683061096252, 0.8164847956978979]


X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Train, 0.4, label = 'Train')
plt.bar(X_axis + 0.2, Test, 0.4, label = 'Test')

plt.xticks(X_axis, X)
plt.xlabel("Models")
plt.ylabel("R2 Score")
plt.title("R2 score of different models")
plt.legend()
plt.show()
```
